```python
import PyPDF2

def extract_text_from_pdf(pdf_file_path):
    # Open the PDF file
    with open(pdf_file_path, 'rb') as file:
        # Create a PDF file reader object
        pdf_reader = PyPDF2.PdfReader(file)
        
        # Initialize an empty string to store extracted text
        extracted_text = ''
        
        # Iterate through each page of the PDF
        for page_num in range(len(pdf_reader.pages)):
            # Get the page object
            page = pdf_reader.pages[page_num]
            
            # Extract text from the page
            text = page.extract_text()
            
            # Append extracted text to the overall text
            extracted_text += text
            
        return extracted_text

# Example usage:
pdf_file_path = 'updated_2english.pdf'
extracted_text = extract_text_from_pdf(pdf_file_path)
print(extracted_text)
```

```python
import cv2
import numpy as np
import pytesseract
from pdf2image import convert_from_path

# Path to the PDF file
pdf_file_path = "/Users/fajrzafar-nu/Documents/Semester 8/FYP 1/Books/English Grade2 Fed.pdf"

# Convert each page of the PDF to an image
pages = convert_from_path(pdf_file_path)

# Initialize an empty list to store extracted text
extracted_text = []

# Iterate over each page image
for i, page in enumerate(pages):
    # Convert the PIL Image object to a numpy array
    page_np = cv2.cvtColor(np.array(page), cv2.COLOR_RGB2BGR)
    
    # Convert the page image to grayscale
    gray_page = cv2.cvtColor(page_np, cv2.COLOR_BGR2GRAY)
    
    # Use Tesseract OCR to extract text from the grayscale image
    text = pytesseract.image_to_string(gray_page)
    
    # Append the extracted text to the list
    extracted_text.append(text)

    # Optionally, you can save the extracted text to a file
    with open(f"page_{i+1}_text.txt", "w") as text_file:
        text_file.write(text)

# Print the extracted text
for i, text in enumerate(extracted_text):
    print(f"Page {i+1} text:")
    print(text)
```
