```python
import cv2
import numpy as np
import os

# Load the image
image_path = "/mnt/data/1725632589104-photo.jpg"  # Your image path
image = cv2.imread(image_path)

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply GaussianBlur to reduce noise and improve contour detection
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Perform edge detection using Canny
edges = cv2.Canny(blurred, 50, 150)

# Find contours in the edge map
contours, _ = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Initialize counter for extracted images
counter = 1

# Set minimum area threshold to filter small contours (you can adjust this value)
min_area = 10000  # Adjust this value as needed based on image size and content

# Loop through the contours and extract the large images
for contour in contours:
    # Get the bounding box for each contour
    x, y, w, h = cv2.boundingRect(contour)
    
    # Calculate the area of the contour
    area = w * h
    
    # Filter based on area to only keep large contours
    if area > min_area:
        # Extract the region of interest (ROI)
        roi = image[y:y+h, x:x+w]
        
        # Save the extracted image
        extracted_image_path = f"/mnt/data/extracted_image_{counter}.jpg"
        cv2.imwrite(extracted_image_path, roi)
        print(f"Saved {extracted_image_path}")
        
        counter += 1

# Display the original image with the bounding boxes (for debugging)
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    area = w * h
    if area > min_area:
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Show the image with the bounding boxes (for debugging)
cv2.imshow("Image with bounding boxes", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
